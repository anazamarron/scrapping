```python
# funcion que convierte un dataframe a json
def convert_float(value):
    value = value.replace("(","")
    value = value.replace(")","")
    value = value.replace(".","")
    value = value.replace(",",".")
    value = value.replace(" ", "0")
    value = value.replace("-","0")
    
    return float(value)
```

```python
#Función que convierte un pdf a texto página a página

def convert_pdf_to_text_by_page(path,pagenum):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    #device = CsvConverter(rsrcmgr, retstr, codec, laparams=laparams)
    device = TextConverter(rsrcmgr, retstr, codec, laparams=laparams)
    fp = file(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0 #is for all
    caching = True
    pagenos=set()
    
    pages = PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password, caching=caching, 
                              check_extractable=True)  
    
    for pageNumber, page in enumerate(pages): #The hell is the first argument?
        totalPages = pageNumber
        if pageNumber == pagenum:
            interpreter.process_page(page)

    fp.close()
    device.close()
    str = retstr.getvalue()
    retstr.close()
    return str, totalPages

```

```python
def traduceMes(mes,lang_origen,lang_destino):
    for index, row in meses.iterrows():
        if(mes.upper().find(row[lang_origen].upper())==0):
            mes = mes.upper().replace(row[lang_origen].upper(),row[lang_destino])
            return mes

def getFechaTime(pdf):
    text,totalPages = convert_pdf_to_text_by_page(pdf,1)
    lista = text.splitlines()
    fecha = ''
    for valor in lista:
        for index, row in meses.iterrows():
            if (valor.upper().find(row["es"].upper())==0 and len(valor)<20):
                valor = valor.upper().replace(row["es"].upper(),row["num"])
                fecha = strptime(valor, '%m %Y')
                break
    return fecha

def getFechaAsJs(pdf):
    jsTime = getFecha(pdf)                
    jstime = int(mktime(jsTime))*1000
    return jstime


def getFecha(pdf):
    strTime = getFechaTime(pdf)    
    strTime = strftime('%B-%Y',strTime)
    strTime = traduceMes(strTime,"en","es")
    return strTime       


def getFechaShort(pdf):
    Time = getFechaTime(pdf)    
    strTime = strftime('%m-%Y',Time)
    Time = datetime.fromtimestamp(mktime(strTime))
    return Time    

def obtenJson(data):
    texto = "[" 
    for index, row in data.iterrows():
        texto += '['
        texto +='\'' + str(index) + '\',' 
        for i in range(len(row)):
            
            texto += str(row[i]) + "," 
            texto += '\'' + str(row[i]) + '\','
        texto = texto[:-1]   
        texto += '],'
        
    texto = texto[:-1]
    texto +=  "]"
    return texto


def obtenJsonComparativa(data):
 
    texto = "[" 
    
    for j, row in data.iterrows():
        
        texto += '['
        for i in range(len(row)):
            texto +='\'' + row.index[i] + '\',' 
        texto = texto[:-1]   
        texto += '],'
        break
            
    for index, row in data.iterrows():
        midatetime = datetime.datetime.fromtimestamp(index)
        midatetime = midatetime.strftime('%m/%Y')
        texto += '['
        texto +='\'' + midatetime + '\','
        for i in range(len(row)):
            texto += str(row[i]) + "," 
        texto = texto[:-1]   
        texto += '],'
        
    texto = texto[:-1]
    texto +=  "]"
    return texto
    
```
